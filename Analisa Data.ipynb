# Analisa Data Siswa SMA yang putus sekolah dengan indeks pembangunan manusia pada periode 2008-2010

```python
import pandas as pd
```

# Deskripsi Data
Data ini merupakan data yang berisikan tentang jumlah siswa SMA yang putus sekolah berdasarkan provinsi pada periode 2008-2011 di Indonesia. Data yang akan dianalisa hanya data pada periode 2008-2010. 

# Melihat Data

```python
data_pddk = pd.read_csv('processed-data-putus-sekolah-sma-2008-2011.csv', delimiter = ',', encoding = "ISO-8859-1")
data_pddk.sort_values(by='jumlah_putus_sekolah',ascending=False)
```

# Memilih kolom

```python
data_pddk.columns
```

# Total keseluruhan siswa SMA yang putus sekolah pada periode 2008-2010 (cakupan nasional)

```python
data_pendidikan= data_pddk[(data_pddk['tahun']<=2010) & (data_pddk['tahun']>=2008)]
print("jumlah keseluruhan siswa SMA yang putus sekolah pada periode 2008-2010 = ",
      data_pendidikan.jumlah_putus_sekolah.sum(), "siswa")
```

# Data siswa SMA yang putus sekolah berdasarkan provinsi pada periode 2008-2010  

```python
data_pendidikan.groupby(['nama_provinsi','tahun']).sum()
```

# Data siswa SMA yang putus sekolah per-tahun pada periode 2008-2010 (cakupan nasional)

```python
total_tahun=data_pendidikan[['jumlah_putus_sekolah','tahun']]
total_year=total_tahun.groupby('tahun').sum()
total_year
```

# Data siswa SMA yang putus sekolah per-provinsi pada periode 2008-2010 

```python
total_provinsi=data_pendidikan[['nama_provinsi','jumlah_putus_sekolah']]
total_prov=total_provinsi.groupby('nama_provinsi').sum()
total_prov.sort_values(by='jumlah_putus_sekolah',ascending=False)
```

# Deskripsi  statistik data siswa SMA yang putus sekolah pada periode 2008-2010 ( cakupan nasional )

```python
print("rata-rata siswa SMA yang putus sekolah pada periode 2008-2010 adalah ",data_pendidikan['jumlah_putus_sekolah'].mean())
print("\nstandar deviasi siswa SMA yang putus sekolah pada periode 2008-2010 adalah ",data_pendidikan['jumlah_putus_sekolah'].std())
print("\njumlah terbesar siswa SMA yang putus sekolah pada periode 2008-2010 adalah" ,data_pendidikan['jumlah_putus_sekolah'].max())
print("\njumlah terkecil siswa SMA yang putus sekolah pada periode 2008-2010 adalah" ,data_pendidikan['jumlah_putus_sekolah'].min())
```

# Data siswa SMA dengan tingkat putus sekolah diatas rata-rata nasional  berdasarkan provinsi per-tahun pada periode 2008-2010

```python
provinsi_atas_rata=data_pendidikan[['nama_provinsi','jumlah_putus_sekolah','tahun']][data_pendidikan[
                    'jumlah_putus_sekolah']>data_pendidikan['jumlah_putus_sekolah'].mean()]
provinsi_atas_rata.groupby(['nama_provinsi','tahun', 'jumlah_putus_sekolah']).sum()
```

# Total keseluruhan siswa SMA diatas rata-rata nasional yang putus sekolah per-satu tahun pada periode 2008-2010

```python
tahun=provinsi_atas_rata.groupby('tahun')
tahun_atas=tahun.sum()
print(tahun_atas)
print("Total keseluruhan siswa SMA diatas rata-rata nasional yang putus sekolah = ",tahun_atas['jumlah_putus_sekolah'].sum())
```

# Data siswa SMA yang putus sekolah diatas rata-rata nasional per-provinsi pada periode 2008-2010

```python
provinsi_atas=data_pendidikan[['nama_provinsi','jumlah_putus_sekolah']][data_pendidikan[
    'jumlah_putus_sekolah']>data_pendidikan['jumlah_putus_sekolah'].mean()]
provinsi_up=provinsi_atas.groupby('nama_provinsi').sum()
provinsi_up.sort_values(by='jumlah_putus_sekolah',ascending=False)
```
